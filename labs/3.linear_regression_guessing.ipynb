# Preparing the data

```python
import numpy as np
import matplotlib.pyplot as plt

from IPython.display import set_matplotlib_formats
set_matplotlib_formats('retina')
%matplotlib inline
```

```python
X = [1.0, 2.0, 3.0]
Y = [2.0, 4.0, 6.0]
```

# Defining the model

```python
def predict(x, w):
    return x * w
```

# Loss function

```python
def loss(y_pred, y):
    return (y_pred - y) * (y_pred - y)
```

# Learning?

```python
w_list = []
mse_list = []

for w in np.arange(0.0, 4.1, 0.1):
    print("w=", w)
    l_sum = 0
    for x_val, y_val in zip(X, Y):
        y_pred_val = predict(x_val, w)
        l = loss(y_pred_val, y_val)
        l_sum += l
        print("\t", x_val, y_val, y_pred_val, l)
    print("MSE=", l_sum / 3)
    w_list.append(w)
    mse_list.append(l_sum / 3)
```

# Results

```python
plt.plot(w_list, mse_list)
plt.ylabel('Loss')
plt.xlabel('w');
```
